**PANDAS**


**Pandas** stands for python data analysis

It stands for open source software library  designed for data manipulation and analysis.

built top of NumPy , efficiently manages large datasets, offering tools for data cleaning , transformation and analysis.

Revolves around two primary data structure : **Series (1D)** and **DataFrame(2D)**


**What is Pandas used for ?**


Reading and writing data from various file formats like CSV, Excel and SQL databases.


Cleaning and preparing data (handling missing values, filtering, removing duplicates).


Merging, joining, and reshaping datasets.


Performing statistical analysis and descriptive statistics.


Visualizing data quickly.


Pandas **Series** is a one-dimensional labeled array that can hold data of any type (integer, float, string, Python objects, etc.). It is similar to a column in an Excel spreadsheet or a database table. 

**Key Features of Pandas Series**

Supports integer-based and label based indexing.


Stores heterogeneous data types.

Offers a variety of built in methods for data manipulation and analysis.



**Creating Pandas Series**

**Position- based Indexing**


```python
import pandas as pd
import numpy as np 

data = np.array(['p','a','l','a','k'])
ser=pd.Series(data)

print(ser[:5])

print(ser[:4])
```

**Label -Based Indexing**

```python
import pandas as pd
import numpy as np

data = np.array(['p','a','l','a','k'])
ser= pd.Series(data,index=[10,11,12,13,14])
print(ser[14])
```

**Binary operation on Pandas Series**




```python
import pandas as pd

ser1 = pd.Series([1, 2, 3], index=['A', 'B', 'C'])
ser2 = pd.Series([4, 5, 6], index=['A', 'B', 'C'])

df_sum = ser1.add(ser2)
print(df_sum)

df_sub = ser1.sub(ser2)
print(df_sum)

df_mul = ser1.mul(ser2)
print(df_mul)

df_div = ser1.div(ser2)
print(df_div)


df_pow = ser1.pow(ser2)
print(df_pow)

```

```python
# importing pandas as pd
import pandas as pd

# Creating the Series
sr = pd.Series([10, 25, 3, 25, 24, 6])

# Create the Index
index_ = ['Coca Cola', 'Sprite', 'Coke', 'Fanta', 'Dew', 'ThumbsUp']

# set the index
sr.index = index_

# Print the series
print(sr)
```

**Example #2: Use Series.mean() function to find the mean of the underlying data in the given series object. The given series object also contains some missing values.**


```python
# importing pandas as pd
import pandas as pd

# Creating the Series
sr = pd.Series([19.5, 16.8, None, 22.78, 16.8, 20.124, None, 18.1002, 19.5])

# Print the series
print(sr)
```

```python
# return the mean
# skip all the missing values
result = sr.mean(skipna = True)

# Print the result
print(result)
```

**Code #1: By default, product of an empty or all-NA Series is 1.**

```python
# importing pandas module 
import pandas as pd 

# min_count = 0 is the default
pd.Series([]).prod()

# When passed  min_count = 1, 
# product of an empty series will be NaN
pd.Series([]).prod(min_count = 1)
```

**Series.abs() method is used to get the absolute numeric value of each element in Series/DataFrame.**

```python
 # importing pandas module 
import pandas as pd 

# creating lists
lst = [2, -10.87, -3.14, 0.12]
lst2 = [-10.87 + 4j]

ser = pd.Series(lst)
ser1 = pd.Series(lst2)

# printing values explaining abs()
print(ser1.abs(), '\n\n', ser.abs())
```

  
**Code #2: Explaining use of abs() on specific row**

```python
# importing pandas module 
import pandas as pd 

df = pd.DataFrame({'Name': ['John', 'Hari', 'Peter', 'Loani'],
                  'Age': [31, 29, 57, 40],
                  'val': [98, 48, -80, -14]})


df['ope'] = (df.val - 87).abs()

df
```

**Conversion Operation on Series**


Conversion operations allow transforming data types within a Series. This can be useful for ensuring consistency in data types. In order to perform conversion operation we have various function which help in conversion like .astype(), .tolist() etc

```python
import pandas as pd

ser = pd.Series([1, 2, 3, 4])
ser = ser.astype(float)
print(ser)
```
